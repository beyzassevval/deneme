```python
#k-means kümeleme
import numpy as no
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans #kütüphaneden k-means i aldık
from sklearn.datasets import make_blobs #veri setini oluşturmak için kullandık-tek tek yazmak yerine benim yerime oluşturucak

#örnek oluşturma
n_sample=300
random_state=42
X,y=make_blobs(n_samples=n_sample,centers=3,cluster_std=2,random_state=random_state)

#veri setini görselleştiricem
plt.scatter(X[:,0],X[:,1],s=50)
plt.title("Veri kümesi")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
#bu veri setini k-means ile eğitelim
n_clusters=3 #küme sayısı
kmeans=KMeans(n_clusters=n_clusters,random_state=random_state,n_init="auto")
y_kmeans=kmeans.fit_predict(X)
#küme merkezlerini al
centers=kmeans.cluster_centers_ #kmeans içerisinde clustercentersiçindekiiler benim küme merkezlerim

#sonuçları görselleştir
plt.scatter(X[:,0],X[:,1],c=y_kmeans,s=50,cmap="viridis")
plt.scatter(centers[:,0],centers[:,1],c="black",s=200,alpha=0.5,label="Küme Merkezleri")
plt.title("Kümeleme Sonuçları")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
